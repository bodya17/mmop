## Завдання 1 (Репліки)

Задана репліка $2^{7-4}$. Походу треба знайти визначальне та генеруюче співвідношення

2 - 2-во рівневий план

7 - максимальна кількість x-сів $x_1, x_2 ...$

7 - 4 = 3 це базові $(x_1, x_2, x_3)$


**$x_i = \pm 1$**

Треба задати 4 $(x_4, x_5, x_6, x_7)$ через базові $(x_1, x_2, x_3)$

### Генеруюче співвідношення

$x_4 = x_1$

$x_5 = x_2$

$x_6 = x_3$

$x_7 = x_1 x_2 x_3$ (це множення)

Може бути заданий парметр "Роздільна здатність"

Роздільна здатність - це найменша кількість множників в правій частині

### Визначальне співвідношення

Множимо праву і ліву частини на то шо є в лівій (зліва отримуємо 1 бо $x_i = \pm 1$)

$1 = x_1x_4$

$1 = x_2x_5$

$1 = x_2x_3x_6$

$1 = x_1x_2x_3x_7$

Тут роздільна здатність = 2

### Тут шукається ефект (хз шо то)
$\hat{\theta_2} \rightarrow \theta_2 + \theta_5 + \theta_{125} + \theta_{236} + \theta_{137}$
